## **Univariate Analysis of Categorical & Continuous Variables**

```python
# --------- Univariate Analysis for Categorical Columns ---------
for col in categorical:
    print(f"\n--- Value Counts for '{col}' ---")
    print(df[col].value_counts())
    
    plt.figure(figsize=(6, 4))
    sns.countplot(x=col, data=df, order=df[col].value_counts().index)
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# --------- Univariate Analysis for Continuous Columns ---------
for col in continuous:
    print(f"\n--- Summary Statistics for '{col}' ---")
    print(df[col].describe())

    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(6, 2))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.tight_layout()
    plt.show()
```

**Observation:**
This section explores the distribution of each individual feature.

- **Categorical Columns:**
  - Uses `countplot` to show the frequency of each category.
  - Helps detect class imbalance (e.g., more employees from one department or job level).

- **Continuous Columns:**
  - Uses `histplot` to visualize distribution and check for skewness.
  - Uses `boxplot` to identify outliers.
  - Descriptive statistics (`describe()`) provide mean, median, min, max, and standard deviation for deeper insights.

This helps in understanding variable behavior and supports decisions on preprocessing (e.g., outlier handling, transformation).


**Bivariate Analysis of Categorical & Continuous Variables** Continuous vs Continuous (Scatterplots)

```python
# -------------------------------
# 1. Continuous vs Continuous
# -------------------------------
print("\n📊 Continuous vs Continuous (Scatterplots):\n")
for i in range(len(continuous)):
    for j in range(i+1, len(continuous)):
        x = continuous[i]
        y = continuous[j]
        plt.figure(figsize=(6, 4))
        sns.scatterplot(x=df[x], y=df[y])
        plt.title(f'Scatterplot: {x} vs {y}')
        plt.tight_layout()
        plt.show()

# -------------------------------
# 2. Categorical vs Continuous
# -------------------------------
print("\n📊 Categorical vs Continuous (Boxplots):\n")
for cat in categorical_cols:
    for con in continuous:
        plt.figure(figsize=(7, 4))
        sns.boxplot(x=df[cat], y=df[con])
        plt.title(f'Boxplot: {con} by {cat}')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

# -------------------------------
# 3. Categorical vs Categorical
# -------------------------------
print("\n📊 Categorical vs Categorical (Countplots with Hue):\n")
for i in range(len(categorical)):
    for j in range(i+1, len(categorical)):
        x = categorical[i]
        hue = categorical[j]
        plt.figure(figsize=(7, 4))
        sns.countplot(x=x, hue=hue, data=df)
        plt.title(f'{x} vs {hue}')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
```

**Observation:**

This section analyzes how two variables relate to each other:

1. **Continuous vs Continuous (Scatterplots):**
   - Scatterplots help identify linear or non-linear relationships, clusters, and trends between numerical variables.
   - Useful for spotting correlations or patterns (e.g., Experience vs Age, SleepHours vs Stress).

2. **Categorical vs Continuous (Boxplots):**
   - Boxplots reveal the distribution of a numerical variable across categories.
   - Ideal for comparing medians, detecting variability, and spotting outliers (e.g., Workload across Departments).

3. **Categorical vs Categorical (Countplots with Hue):**
   - Visualizes how two categorical variables intersect.
   - Helps understand class distribution overlaps and imbalances (e.g., Gender vs EmpType).

Such visual comparisons are key to uncovering patterns, dependencies, or anomalies before modeling.
